### Zadanie 2 
Napisz funkcje, która dla dowolnej liczby zmiennoprzecinkowej w precyzji 64-bitowej (w standardzie IEEE 754) przedstawi w wersji 
dziesiętnej i binarnej składniki liczby: znak, wykładnik i mantysa. Można wykorzystać między innymi bibliotekę bitstring. Zbadaj jak reprezentowaną są poszczególne składniki tego formatu: zakres, sposób kodowania liczby etc. 

```python
import bitstring as bs

def toDec(x, p):
    res = 0
    for i in x[0:]:
        if i == '1':
            res += 2 ** p
        p -= 1
    return res

def num_info(number):
    bf = bs.BitArray(float=number, length=64)
    
    print 'Składniki liczby w postaci binarnej'
    print 'Znak:', bf.bin[0]
    print 'Cecha:', bf.bin[1:12]
    mantysa = '1.'+bf.bin[12:64]
    print 'Mantysa:', mantysa, '\n'
    
    print 'Składniki liczby w postaci dziesiętnej'
    print 'Znak: ', '+' if bf.bin[0]=='0' else '-'
    print 'Cecha:', int(bf.bin[1:12], 2)
    print 'Mantysa:', 1+toDec(bf.bin[12:64], -1)    
    
num_info(110.375)
```

### Zadanie 5
Sprawdź czym jest postać znormalizowana liczby typu float. Poniższy program generuje ciąg liczb. Wyświetl ich postać binarną i sprawdź czy są w postaci znormalizowanej (napisz stosowny kod, który to sprawdza). 

Postać **_znormalizowana_** liczby typu float to taka postać, w której mantysa reprezentuje liczbę w postaci 1,xxx...xxx, a nie 0,xxx...xxxx

```python
def decimalToBinary(num): 
  
    binary = ""    
    integral = int(num)  
    fractional = num - integral
    
    while (integral) :   
        rem = integral % 2 
        binary += str(rem);  
        integral //= 2
    
    if integral == 0:
        binary = '0'
    else:
        binary = binary[ : : -1]  
    
    binary += '.'
    
    while (fractional > 0): 
        fractional *= 2
        fract_bit = int(fractional)  
        
        if (fract_bit == 1):  
            fractional -= fract_bit  
            binary += '1'
        else : 
            binary += '0'
            
    return binary

def print_info(a, b):
    print a, b, 'znormalizowana' if b[0] == 0 and b[1] == '.' else 'nieznormalizowana'

a=np.float32(1.1)
for i in range(0, 149):
    a=a/np.float32(2.0)
    b = decimalToBinary(a)
    print_info(a,b)
    
    
```

Porównać reprezentację bitową liczby 1/3 dla typów float o różnej precyzji: float16, float32, float64. Sprawdź co się dzieje gdy zmienna o mniejszej precyzji konwertowana jest do wyższej. 

```python
import bitstring

f16 = (np.float16)(1.0/3.0)
print 'reprezentacja dziediętna:', f16, ', binarna:', decimalToBinary(f16)

f32 = (np.float32)(1.0/3.0)
print 'reprezentacja dziediętna:', f32, ', binarna:', decimalToBinary(f32)

f64 = (np.float64)(1.0/3.0)
print 'reprezentacja dziediętna:', f64, ', binarna:', decimalToBinary(f64)

f = (np.float64)(f32)
print 'reprezentacja binarna float32 skonwertowanego do float64:', '0.' + bitstring.BitArray(float=f, length=64).bin
```

Jak widzimy przy konwertowaniu float'a o mniejszej precyzji do wyższej dodatkowe bity, który pojawiły się przy zwiększeniu precyzji są zerami. Więc nie ma sensu konwertować z mniejszej precyzjido wyższej.

Poniższy program nie działą prawidłowo, bo przy mnożeniu zaokrąglenie liczby jest wykonywane jeden raz, a przy wykonywaniu operacji dodawania w pętli N razy.

```python
N = 100000000
C = 0.09531258654533566;

sum = (np.float128)(0.0)
for i in range(0, N):
    sum += C
    
print(sum)
print(N*C)
```

Dla sumowania liczb można stosować algorytm _Kahana_, który daje dokłądniejszy wynik. Algorytm Kahana jest przedstawiony poniżej

```python
N = 100000000
C = 0.09531258654533566;

def kahansum(nbr, input):
    summ = c = 0
    for i in range(0, input):
        nbr_ = nbr - c
        sum_ = summ + nbr_
        c = (sum_ - summ) - nbr_
        summ = sum_
    return summ

print(kahansum(C, N))
print(N*C)

```
